```python
import cPickle
```

```python
def unpickle(batch_file):
    fo = open(batch_file, 'rb')
    dictionary = cPickle.load(fo)
    fo.close()
    return dictionary
```

Each of the batch_files are pickled. We use the unpickle function above, which returns the batch_file in a dictionary format. Each batch_file contains 10,000 images:

### Use sklearn logistic regression, just to see how well it would perform out of the box

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# my X
X_lr = data_batch_1['data']
```

```python
# my Y
y_lr = data_batch_1['labels']
```

```python
x_train_lr, x_test_lr, y_train_lr, y_test_lr = train_test_split(X_lr, y_lr, random_state=42)
```

```python
x_train_lr = x_train_lr.astype('float64')
x_test_lr = x_test_lr.astype('float64')
```

```python
sc = StandardScaler()
sc.fit(x_train_lr)
x_train_lr = sc.transform(x_train_lr)
x_test_lr = sc.transform(x_test_lr)
```

```python
lr = LogisticRegression(penalty='l2', multi_class='multinomial', solver='sag', n_jobs=-1)
```

```python
lr.fit(x_train_lr, y_train_lr)
```

```python
train_score = lr.score(x_train_lr, y_train_lr)
test_score = lr.score(x_test_lr, y_test_lr)
print "train_score: ", train_score
print "test_score: ", test_score
```
